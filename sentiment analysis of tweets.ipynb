```python
#cleaning  the text 
def nlp_preprocessing(total_text, index, column):
    if type(total_text) is not int:
        string = ""
        total_text = re.sub('((www\.[^\s]+)|(https?://[^\s]+))','',total_text)
        # replace every special char with space
        total_text = re.sub('[^a-zA-Z0-9\n]', ' ', total_text)
        # replace multiple spaces with single space
        total_text = re.sub('\s+',' ', total_text)
        # converting all the chars into lower-case.
        total_text = total_text.lower()
        
        df[column][index] = total_text
```

```python
import time
from datetime import datetime
start_time = time.time()
for index, row in df.iterrows():
    if type(row['Tweets']) is str:
        nlp_preprocessing(row['Tweets'], index, 'Tweets')
    else:
        print("there is no text description for id:",index)
print('Time took for preprocessing the text :',time.time() - start_time, "seconds")
```

```python
# for finding the sentiment we are using nltk sentiment analyser
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
sid = SentimentIntensityAnalyzer()
neg = []
neu = list()
pos = list()
a = []
compound = list()
for index, row in df.iterrows():
    ss = sid.polarity_scores(df["Tweets"][index])
    neg.append(ss['neg'])
    neu.append(ss["neu"])
    pos.append(ss["pos"])
    compound.append(ss["compound"])
    a.append(index)
    
df_1 = pd.DataFrame(list(zip(neg, neu, pos, compound)), index = a,
               columns =['neg', 'neu', 'pos', 'compound'])   
```

```python
df_1
```

```python
d= []
for index, row in df_1.iterrows():
    if df_1["neg"][index] > df_1["pos"][index]:
        d.append("negetive")
    elif df_1["neg"][index] == df_1["pos"][index]:
        d.append("neutral")
    else:
        d.append("positive")
```

```python
d
```

```python
df_1["sentiment"] = d
```

```python
df_1
```

```python
df = df.join(df_1)
```
